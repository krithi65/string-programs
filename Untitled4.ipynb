32. find the number of strings that are an anagram of 'kabali'.If there exists no anagram for the given string print '0'.
Input Size : 1 <= N <= 1000
Sample Testcase :
INPUT
5
kabali
kaabli
kababa
kab
kabail
OUTPUT
3

```python

n = int(input())
l = []
for i in range(n):
    l.append(str(input()))
ref = 'kabali'
count = 0
for i in l:
    flag = True
    for j in i:
        if i.count(j) != ref.count(j):
            flag = False
            break
    if flag:
        count+=1
print(count)
```

45.Given a sentence interchange the between the word 'and'.
Input Size : |S| <= 1000000
Sample Testcase :
INPUT
jack and jill went up and down to get water
OUTPUT
jill and jack went down and up to get water

```python
l = input().split(" ")
a=[]
for i in range(len(l)):
    if l[i] == "and":
        (l[i+1], l[i-1]) = (l[i-1], l[i+1])
for i in l:
    a.append(i)
print(*a)
```

49.print reverse all words except the first and last words.

```python
s = input().split(" ")
a=[]
for i in range(len(s)):
    if i == 0:
        l1=s[i]
        a.append(l1)
    elif i == (len(s)-1):
        l2=s[i]
        a.append(l2)
    else:
        l3=s[i][::-1]
        a.append(l3)
print(*a)        
```
